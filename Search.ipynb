```python
# Binary search
def binary_search(arr, target):
    low, high = 0, len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1
binary_search([1, 2, 3, 4, 5], 3)
```

```python
#Linear Search
def linear_search(arr, target): 
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1
linear_search([1, 2, 3, 4, 5], 3)
```
